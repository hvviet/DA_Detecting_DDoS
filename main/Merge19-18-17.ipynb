```python
import numpy as np
from dask import dataframe as pd
```

```python
DrDoS_DNS_data_1_per = pd.read_csv('../datasets/DrDoS_DNS_data_1_per.csv',dtype={'SimillarHTTP': str})
DrDoS_LDAP_data_2_0_per = pd.read_csv('../datasets/DrDoS_LDAP_data_2_0_per.csv',dtype={'SimillarHTTP': str})
DrDoS_MSSQL_data_1_3_per = pd.read_csv('../datasets/DrDoS_MSSQL_data_1_3_per.csv',dtype={'SimillarHTTP': str})
DrDoS_NetBIOS_data_1_3_per = pd.read_csv('../datasets/DrDoS_NetBIOS_data_1_3_per.csv',dtype={'SimillarHTTP': str})
DrDoS_NTP_data_data_5_per = pd.read_csv('../datasets/DrDoS_NTP_data_data_5_per.csv',dtype={'SimillarHTTP': str})
DrDoS_SNMP_data_1_3_per = pd.read_csv('../datasets/DrDoS_SNMP_data_1_3_per.csv',dtype={'SimillarHTTP': str})
DrDoS_SSDP_data_2_per = pd.read_csv('../datasets/DrDoS_SSDP_data_2_per.csv',dtype={'SimillarHTTP': str})
DrDoS_UDP_data_2_per = pd.read_csv('../datasets/DrDoS_UDP_data_2_per.csv',dtype={'SimillarHTTP': str})
Syn_data_4_per = pd.read_csv('../datasets/syn_data.csv',dtype={'SimillarHTTP': str})
UDPLag_data_2_0_per = pd.read_csv('../datasets/UDPLag_data_2_0_per.csv',dtype={'SimillarHTTP': str})
```
